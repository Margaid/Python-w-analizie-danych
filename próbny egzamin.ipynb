# E 
Napisz funkcjÄ™, ktÃ³rej zadaniem bÄ™dzie wyÅ›wietlenie maksymalnej bÄ…dÅº minimalnej wartoÅ›ci zadanej kolumny z pobranej ramki danych. Ramka danych i nazwa kolumny powinny byÄ‡ parametrami funkcji.

```python
def min_or_max(df,col,m):
    if m=='min':
        return df[df[col]==df[col].min()]
    if m=='max':
        return df[df[col]==df[col].max()]
    else:
        return 'WpisaÅ‚eÅ› bÅ‚Ä™dny parametr'
    
min_or_max(states,('Flag, name andpostal abbreviation [1]','Flag, name andpostal abbreviation [1]'),'min')
```
