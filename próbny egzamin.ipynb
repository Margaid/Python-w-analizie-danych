# Zadanie 2
A. Wyświetl stany, które mają najmniej i najwięcej reprezentantów do Kongresu (ostatnia kolumna). Napisane polecenia nie powinny zawierać sztywnych wartości.

```python
number_of_rep_max=states[states.iloc[:,-1]==states.iloc[:,-1].max()]
number_of_rep_max.iloc[:,[0,-1]]
```

```python
number_of_rep_min=states[states.iloc[:,-1]==states.iloc[:,-1].min()]
number_of_rep_min.iloc[:,[0,-1]]
```

# B.
Dla dziesięciu najbadziej zaludnionych stanów pokaż na wykresie ilu mają reprezentantów.

```python
sorted_by_population=states.sort_values([('Population[3]','Population[3]')],ascending=False).iloc[:10,:]
sorted_by_population
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.bar(sorted_by_population[('Flag, name andpostal abbreviation [1]','Flag, name andpostal abbreviation [1]')],
        sorted_by_population[('Numberof Reps.','Numberof Reps.')], color='green')
plt.xticks(np.arange(0,10),rotation=70)
plt.title("Ilość reprezentantów w 10 najbardziej zaludnionych stanach")
plt.ylabel("ilość")
plt.show()
```

# C. 
Zobrazuj gęstośc zaludnienia (Population/Land area) poszczególnych stanów.

```python
density_of_population=states[('Population[3]','Population[3]')]/states[('Land area[4]','km2')]
density_of_population
```

```python
plt.figure(figsize=(18,10)) 
plt.bar(states[('Flag, name andpostal abbreviation [1]','Flag, name andpostal abbreviation [1]')],
       density_of_population)
plt.xticks(rotation=80)
plt.title("Gęstość zaludnienia wszystkich stanów")
plt.ylabel('gęstość')
plt.grid()
plt.show()
```

# E 
Napisz funkcję, której zadaniem będzie wyświetlenie maksymalnej bądź minimalnej wartości zadanej kolumny z pobranej ramki danych. Ramka danych i nazwa kolumny powinny być parametrami funkcji.

```python
def min_or_max(df,col,m):
    if m=='min':
        return df[df[col]==df[col].min()]
    if m=='max':
        return df[df[col]==df[col].max()]
    else:
        return 'Wpisałeś błędny parametr'
    
min_or_max(states,('Flag, name andpostal abbreviation [1]','Flag, name andpostal abbreviation [1]'),'min')
```

```python
min_or_max(states,('Established[A]','Established[A]'),'max')
```
